```python
import numpy as np
```

```python
y = np.array([3.0,5.0,9.0,10.0])
```

```python
A = np.array([
    [1,0,1],
    [1,2,3],
    [1,4,8],
    [1,6,10],
    ])
```

```python
# Three ways to computed beta
# Use matrix inverse. Never do this. Very slow of doing the inverse of the entire thing
beta = np.linalg.inv((A.T.dot(A))).dot(A.T.dot(y))

# To solve. This is more efficient. It is still not good. You don't actually need to use A.T
beta = np.linalg.solve(A.T.dot(A), A.T.dot(y))

# Use a less square function. This is the fastest
beta = np.linalg.lstsq(A,y,rcond=None)[0]

beta
```

# Non Linear Transformation

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 100)
y = x * np.exp(-x)

plt.plot(x, y, 'o')
```

```python
def transform(x, d):
    """
    Creates the transform data matrix
    In this case, we create a matrix Z[:,i] = x ** i
    i = 0, ... d
    for polynomial fitting
    """
    n = len(x)
    Z = np.zeros((n, d))
    for i in range(d):
        Z[:,i] = x ** i
    return Z
```

```python
Z = transform(x, len(x))
```
